```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [1,2,3]
```

```python
plt.plot(data)
plt.show() # in the terminal as required to displsy the chart
```

```python
plt.plot(data , linewidth = 10)
plt.title('example' , fontsize = 15)
plt.xlabel('X-axes' , fontsize = 15)
plt.ylabel('Y-axes' , fontsize = 15)
```

```python
data = [1,2,3]
plt.plot(data , linewidth = 5)
plt.title('random number' , fontsize = 15)
plt.xlabel('X-axes' , fontsize = 15)
plt.ylabel('Y-axes' , fontsize = 15)
plt.show()
```
