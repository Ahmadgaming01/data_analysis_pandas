```python
import pandas as pd 
import matplotlib.pyplot as plt
 
```

```python
df = pd.read_csv('cars.csv')
df.head()
df.rename(
    columns={
        'Name':'name',
        'Malfunctions' :'malfunctions',
        'age':'age',

    } ,
    inplace=True
)
y1 = df['age']
y2 = df['malfunctions']


plt.subplot(1 , 2 ,1)
plt.plot(y1 ,y2 , marker = 'o')

plt.subplot(1,2,2)
plt.plot(y1 ,y2 , marker = 'o')

plt.show()
```

```python
plt.subplot(2 , 1 ,1)
plt.plot(y1 ,y2 , marker = 'o')

plt.subplot(2,1,2)
plt.plot(y1 ,y2 , marker = 'o')

plt.show()
```
