```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y = [35 , 25 , 25 , 15]
plt.pie(y)
plt.show()
```

```python
maylabels = ['apple' , 'bananas' , 'cherries' , 'dates']
plt.pie(y , labels = maylabels)
plt.show()
```

```python
myexplode = [0.2 , 0 ,0,0]
plt.pie(y , labels=maylabels ,explode=myexplode)
```

```python
myexplode = [0.2 , 0 ,0,0]
plt.pie(y , labels=maylabels ,explode=myexplode , shadow=True)
```

```python
plt.pie(y , labels= maylabels)
plt.legend()
```
