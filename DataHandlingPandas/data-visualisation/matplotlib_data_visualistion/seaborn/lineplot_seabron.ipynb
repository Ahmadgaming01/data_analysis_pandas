```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('euro-daily-hist_1999_2022.csv')
df.rename(columns={
    'Date': 'date',
    'Australian dollar':'Australian_dollar',
    'Canadian dollar':'Canadian_dollar',
    'Swiss franc':'Swiss_franc'
} , inplace=True)
```

```python
df = df.iloc[: , [0,1,4,-2]]
df.columns = ['date' , 'Australian_dollar', 'Canadian_dollar' , 'Swiss_franc']
print(df)
```

```python
df = pd.melt(df , id_vars='date' , value_vars=['Australian_dollar', 'Canadian_dollar' , 'Swiss_franc'] , value_name='Euro rate' , var_name='Currency')
df['date'] = pd.to_datetime(df['date'])

df = df[df['date'] >= '2023-12-01'].reset_index(drop = True)

df['Euro rate'] = pd.to_numeric(df['Euro rate'])

print(f'Currencies:{df.Currency.unique()}/n')

print(df)

```

```python
usd = df[df['Currency']=='Australian_dollar'].reset_index(drop= True)
sns.lineplot(x='date' , y='Euro rate', data = usd)
```

```python
fig = plt.subplots(figsize = (20,5))
sns.lineplot(x = 'date' , y = 'Euro rate' , data = usd).set(title  = ' Euro-USD' , xlabel = 'Date' , ylabel = 'Euro')
```

```python
fig = plt.subplots(figsize = (20,5))
sns.lineplot(x = 'date' , y = 'Euro rate' , data = usd , linestyle = 'dotted' , linewidth = 5 ).set(title  = ' Euro-USD' , xlabel = 'Date' , ylabel = 'Euro')
```

```python
fig = plt.subplots(figsize = (20,5))
sns.lineplot(x = 'date' , y = 'Euro rate' , data = usd , marker = '*'  , markerfacecolor = 'limegreen' , markersize = 20).set(title  = ' Euro-USD' , xlabel = 'Date' , ylabel = 'Euro')
```
